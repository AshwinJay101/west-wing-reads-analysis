```python
import pandas as pd
data = pd.read_csv('./Data/Outputdata.csv')
data_text = data[['content']]
data_text['index'] = data_text.index
print(len(data_text))
print(data_text['content'][0])
```

```python
data_text = data_text[data_text['content'].notnull()]
bow=[string.split() for string in data_text['content']]
print(bow)
```

```python
import gensim
dictionary = gensim.corpora.Dictionary(bow)
print(dictionary)
```

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in bow]
print(bow_corpus)
```

```python
from gensim import corpora, models
tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
from pprint import pprint
for doc in corpus_tfidf:
    pprint(doc)
    break
```

```python
len(corpus_tfidf)
```

```python
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, num_topics=13, id2word=dictionary, passes=50, iterations =100)
#print(lda_model_tfidf.print_topics())
for idx, topic in lda_model_tfidf.print_topics():
    print('\nTopic: {} \nWord: {}'.format(idx, topic))
```

```python
test_data = pd.read_csv('./Data/Testdata.csv')
test_data_text = test_data[['content']]
test_data_text = test_data_text[test_data_text['content'].notnull()]
test_bow=[string.split() for string in test_data_text['content']]
for i in range(len(test_bow)):
    unseen_document = test_bow[i]
    print("\n\nDocument"+str(i)+"\n")
    #print(unseen_document)
    bow_vector = dictionary.doc2bow(unseen_document)
    for index, score in sorted(lda_model_tfidf[bow_vector], key=lambda tup: -1*tup[1]):
        print("Score: {}\t Topic: {} {}".format(score, index, lda_model_tfidf.print_topic(index, 5)))
```
